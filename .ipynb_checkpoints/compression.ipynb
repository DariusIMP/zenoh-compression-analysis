# Compression Stats

In this notebook we want to compare the behavior of a compression codec for Zenoh in terms of time cost and compression ratio.

Compression results vary depending on the size of the original message as well as its complexity; a series of bytes with high entropy or complexity is hard to compress, and the gain in size may not justify the time penalty the compression induces, in fact the size of the compression result may even be bigger than its input. 


## Sources

To get these measurements below, we ran benches varying the size of the batches to compress and their complexity.

The sizes correspond to 1, 2, 4, 8, 16, 24, 32, 40, 56 and 64 kilobytes (the max size of a zenoh batch)

The batches are filled with data from files whose entropy/complexity varies.

## Operations

* Simple encoding: consists of basically encoding the batches as vectors with the Zenoh060 codec.

* Compression: consists of 3 steps:
    * Encoding the input with Zenoh060 into a vector of u8
    * Compressing the output
    * Encoding again with Zenoh060 as a vector of u8

The time metrics were obtained using cargo-bench. Cargo bench runs a piece of code a certain amount of times, measures the time it took to execute it each time and makes an average with other metrics such as the invariance. In this notebook we only used the time averages.

The size metrics were obtained using the cargo-bench command as well but it does not consist of a benchmark in the proper sense. We simply used cargo bench to be able to execute a piece of code that would make a compression for every specified size and every specified complexity, on both modalities (simple encoding and compression).
These measurements were then stored into a csv file we later load in this notebook.

```python
import os
import json
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
KB = 1024
SIZES = [1 * KB, 2 * KB, 4 * KB, 8 * KB, 16 * KB, 24 * KB, 32 * KB, 40 * KB, 48 * KB, 56 * KB, 64 * KB]
ENTROPIES = ["LOW", "MIDDLE", "HIGH"]
```

```python
def get_estimates_json_file_path(size: int, entropy: str, compression: bool) -> str:
    """Retrieves the filepath of the JSON file containing the metrics for that combination of inputs."""
    header = "Simple encoding"
    if compression:
        header = "Compression"
    return "../criterion/Compression/" + header + " - " + entropy + " entropy/" + str(size) + "/new/estimates.json"
```

```python
def load_time_averages_ns() -> pandas.DataFrame:
    size_col = []
    entropy_col = []
    compression = []
    avg_time_ns = []
    for size in SIZES:
        for entropy in ENTROPIES:
            for is_compressed in [False, True]:
                size_col += [size]
                entropy_col += [entropy]
                compression += [is_compressed]
                with open(get_estimates_json_file_path(size, entropy, is_compressed), 'r') as f:
                    data = json.load(f)
                    avg_time_ns += [data["mean"]["point_estimate"]]
    return pandas.DataFrame(data={"size" : size_col, "entropy": entropy_col, "is_compressed": compression, "avg_time_ns": avg_time_ns})
```

```python
def load_compression_stats() -> pandas.DataFrame:
    return pandas.read_csv("../metrics.csv")
```

```python
times_df = load_time_averages_ns()
sizes_df = load_compression_stats()
df = pandas.merge(times_df, sizes_df, on=["size", "is_compressed", "entropy"])

df["avg_time_us"] = df["avg_time_ns"] / 1000
df["size_kb"] = df["size"]/KB
df["output_size_kb"] = round(df["output_size"]/KB, 2)
df["compression_ratio"] = round(df["output_size"] / df["size"], 3)
```

# Compression results

Compressing the batches consumes time, but what's the real payload of a compressed batch compared to the payload of a non-compressed batch? 

How much time does it take us to send the same amount of bytes in a simple way compared to a compressed way

## Compression size analysis

How many bytes do we save by compressing the batches depending on the complexity and the size of the samples

```python
def plot_compression_sizes_bar_chart(entropy: str):
    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, entropy)).get(["size_kb", "output_size_kb"])

    x = np.arange(len(sizes_df))  # the label locations
    width = 0.3  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    offset = width * multiplier
    rects = ax.bar(x + offset, sizes_df.get("size_kb"), width, label="Input size [KB]")
    ax.bar_label(rects, padding=3)
    multiplier += 1
    offset = width * multiplier
    rects = ax.bar(x + offset, sizes_df.get("output_size_kb"), width, label="Compressed size [KB]")
    ax.bar_label(rects, padding=3)


    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.grid(True)
    ax.set_ylabel('Output size [KB]')
    ax.set_xlabel('Batch size [KB]')
    ax.set_title('Batch sizes before and after compression for **' + entropy + '** complexity batches in Kilo Bytes')
    ax.set_xticks(x + width, sizes_df.get("size_kb"))
    ax.legend(loc='upper left', ncols=2)
    fig.set_figwidth(15)
    plt.show()
```

```python
plot_compression_sizes_bar_chart("LOW")
```

```python
plot_compression_sizes_bar_chart("MIDDLE")
```

```python
plot_compression_sizes_bar_chart("HIGH")
```

```python
def plot_compression_ratio():
    fig, ax = plt.subplots()

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "LOW")).get(["size_kb", "compression_ratio"])
    ax.plot(sizes_df["size_kb"], sizes_df["compression_ratio"], label="Low complexity", marker='o')

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "MIDDLE")).get(["size_kb", "compression_ratio"])
    ax.plot(sizes_df["size_kb"], sizes_df["compression_ratio"], label="Middle complexity", marker='o')

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "HIGH")).get(["size_kb", "compression_ratio"])
    ax.plot(sizes_df["size_kb"], sizes_df["compression_ratio"], label="High complexity", marker='o')

    ax.set_xticks(sizes_df["size_kb"])
    ax.set_ylabel('Compression ratio')
    ax.set_xlabel('Batch size [KB]')
    ax.set_title('Compression ratio depending on the size and complexity of the batches.')
    ax.grid(True)
    ax.legend(loc='center right', ncols=1)
    fig.set_figwidth(15)

    # ax.tick_params(labelcolor='r', labelsize='medium', width=3)
    plt.show()
```

```python
plot_compression_ratio()
```

Depending on the complexity/entropy of the batch and on the size, it may not even be worth it to compress it. For the high complexity batches, independently of its size, the result is even worse than without compression.
This is particularly concerning for 64KB batches since the maximum size of a batch in Zenoh is 64KB. If we ever attempt to compress a max size batch and its entropy is high enough so that the end result of the compression is bigger than it's original size, then the result won't fit in a batch.

Besides that, it's naive to assume we'll mostly deal with low complexity batches, instead it's more reasonable to look at the middle complexity batches results.
In the plot above, we see the compression ratio for a middle complexity payload stabilizes around 0.4 after 16KB. However what's the time it consumes to achieve that compression?

# Time analysis

## Average

```python
def plot_compression_avg_time_us():
    fig, ax = plt.subplots()

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "LOW")).get(["size_kb", "avg_time_us"])
    ax.plot(sizes_df["size_kb"], sizes_df["avg_time_us"], label="Low complexity", marker='o')

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "MIDDLE")).get(["size_kb", "avg_time_us"])
    ax.plot(sizes_df["size_kb"], sizes_df["avg_time_us"], label="Middle complexity", marker='o')

    sizes_df = df.groupby(["is_compressed", "entropy"]).get_group((True, "HIGH")).get(["size_kb", "avg_time_us"])
    ax.plot(sizes_df["size_kb"], sizes_df["avg_time_us"], label="High complexity", marker='o')

    ax.set_xticks(sizes_df["size_kb"])
    ax.set_ylabel('Average time [μs]')
    ax.set_xlabel('Batch size [KB]')
    ax.set_title('Average time for compression in micro seconds')
    ax.grid(True)
    ax.legend(loc='upper left', ncols=1)
    fig.set_figwidth(15)

    # ax.tick_params(labelcolor='r', labelsize='medium', width=3)
    plt.show()
```

What's the average time the compression takes depending on the batch size?

```python
plot_compression_avg_time_us()
```

The results are pretty strange but we don't fully know the internal working of the compression algorithm. One may expect that a high complexity batch would consume more time, but the time it takes is similar to the low complexity batch, and the middle complexity batch is the one whose average time increases linearily. It's curious however how at 64KB that average time for the middle complexity batch decreases to 17 micro seconds. However the [benches from lz4_flex](https://github.com/PSeitz/lz4_flex/tree/a122673bf8349dc5554551ad24a7280dbd60caca#benchmarks) have a similar behavior at 64KB.

## Ratio

How many extra times does it take to compress the batches compared to not doing so?

```python
def get_entropy_times_dataframe(entropy: str) -> pandas.DataFrame:
    entropy_group = df.groupby(["entropy"]).get_group(entropy)
    compressed_avg_time_ns = entropy_group[entropy_group["is_compressed"] == True]
    simple_avg_time_ns = entropy_group[entropy_group["is_compressed"] == False]
    entropy_times_df = pandas.DataFrame(data={"size" : compressed_avg_time_ns["size"].values, "compressed_avg_time_ns": compressed_avg_time_ns["avg_time_ns"].values, "simple_avg_time_ns": simple_avg_time_ns["avg_time_ns"].values})
    entropy_times_df["time_ratio"] = entropy_times_df["compressed_avg_time_ns"]/entropy_times_df["simple_avg_time_ns"]
    return entropy_times_df

low_entropy_times_df = get_entropy_times_dataframe("LOW")
mid_entropy_times_df = get_entropy_times_dataframe("MIDDLE")
high_entropy_times_df = get_entropy_times_dataframe("HIGH")
```

```python
def plot_time_compression_ratio():
    fig, ax = plt.subplots()

    sizes_kb = list(map(lambda size: int(size / KB), low_entropy_times_df["size"]))
    ax.plot(sizes_kb, low_entropy_times_df["time_ratio"], label="Low complexity", marker='o')
    ax.plot(sizes_kb, mid_entropy_times_df["time_ratio"], label="Middle complexity", marker='o')
    ax.plot(sizes_kb, high_entropy_times_df["time_ratio"], label="High complexity", marker='o')

    ax.set_xticks(sizes_kb)
    ax.set_ylabel('Time ratio')
    ax.set_xlabel('Batch size [KB]')
    ax.set_title('Compression / Simple encoding time ratio')
    ax.grid(True)
    ax.legend(loc='upper right', ncols=1)
    fig.set_figwidth(15)

    plt.show()
```

```python
plot_time_compression_ratio()
```

Surely, compressing takes a significant amount of time compared to not encoding, particularly for small batches.
When analyzing the compression ratio in the previous section, we saw that for middle complexity batches, the compression ratio was of around 0.4 after 16KB. However in this plot above we see that achieving such results takes **40 times** more than without compressing it. If we make the maths for a batch of 32 KB for instance, `32 * 40 = 1280`, meaning that for the same time it would take us to compress a 32KB, we could have encoded 1280KBs of data corresponding to 40 batches.

The routers will pay the cost in time and processing of these operations. It's very likely that the routers could become a bottleneck in the communication, it'll probably cause a spike in latency.
